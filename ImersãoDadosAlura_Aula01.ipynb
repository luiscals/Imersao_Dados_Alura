```
#importando bibliotecas de visualização
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Fazendo um gráfico de dispersão de duas notas

plt.figure(figsize=(12,6))
ax = sns.scatterplot(x=dados.NU_NOTA_LC, y=dados.NU_NOTA_REDACAO, hue=dados.IN_TREINEIRO)
ax.set_title('Dispersão entre as notas de redação e Linguagem e Código com treineiros');

```

```
#CROSSTAB TREINEIRO/IDADE
pd.crosstab(index=dados['NU_IDADE'], columns=dados['IN_TREINEIRO'])
```

```
#idades dos treineiros
dados.query('IN_TREINEIRO == 1')['NU_IDADE'].value_counts().sort_index()
```

```
#usando subset
treineiros = dados[dados['IN_TREINEIRO'] == 1]
candidatos = dados[dados['IN_TREINEIRO'] == 0]
```

```
#Criando dois histogramas juntos

plt.figure(figsize=(12,6))
sns.kdeplot(x=treineiros['NU_IDADE'], color='r');
sns.kdeplot(x=candidatos['NU_IDADE'], color='b');
plt.title('KdePlot de Idades por Treineiros e Não treineiros', fontsize=16);
plt.legend(labels=['Treineiros', 'Não Treineiros'])
plt.show()

```
